# CRIMES IN CHICAGO

 • The Crimes in Chicago dataset, sourced from the Chicago Police Department's 
CLEAR system, spans incidents reported from 2001 to the present, excluding the 
latest seven days. With over 6 million records

• It offers key details like unique identifiers, case numbers, dates, and arrest

and city area evolution. Gratitude to the City of Chicago and the Chicago Police 
Department for openly sharing this invaluable dataset, fostering insights into the 
dynamic landscape of crime in the city.

# Crime Trends Over Time (Yearly and Monthly)

```python
# Yearly Trends of Overall Crime
# Group by year to count the number of crimes per year
crime_by_year = df.groupby('Year').size()

# Plot yearly trends
plt.figure(figsize=(10, 6))
sns.lineplot(x=crime_by_year.index, y=crime_by_year.values)
plt.title('Total Crime in Chicago (2001 - Present)')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# Yearly Trends for Different Types of Crime (e.g., theft, assault, burglary)
# Filter for specific types of crime
crime_types = ['THEFT', 'ASSAULT', 'BURGLARY']
df_filtered = df[df['Primary Type'].isin(crime_types)]

# Group by year and type of crime
crime_type_by_year = df_filtered.groupby(['Year', 'Primary Type']).size().unstack()

# Plot the trends
plt.figure(figsize=(10, 6))
crime_type_by_year.plot()
plt.title('Yearly Trends of Theft, Assault, and Burglary')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Type')
plt.show()
3
```

```python
# Seasonal Trends (Monthly)
# Group by year and month to calculate the number of crimes per month
crime_by_month = df.groupby(['Year', 'Month']).size().unstack()

# Plot the monthly trends
plt.figure(figsize=(10, 6))
sns.heatmap(crime_by_month, cmap="YlGnBu", cbar=True)
plt.title('Seasonal Crime Pattern (Monthly)')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```

# Geographical Analysis

```python
#Police Beats or Districts with Highest Crime Incidence
# Group by district and count the number of crimes
crime_by_district = df.groupby('District').size().sort_values(ascending=False)

# Plot top 10 districts
plt.figure(figsize=(10, 6))
sns.barplot(x=crime_by_district.index[:10], y=crime_by_district.values[:10], palette='viridis')
plt.title('Top 10 Districts with Highest Crime Incidence')
plt.xlabel('District')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# Violent vs Property Crimes in Different Community Areas
# Filter for violent and property crimes
violent_crimes = ['ASSAULT', 'BATTERY', 'HOMICIDE', 'ROBBERY']
property_crimes = ['BURGLARY', 'THEFT', 'MOTOR VEHICLE THEFT']

df['Crime Type'] = df['Primary Type'].apply(lambda x: 'Violent' if x in violent_crimes else 'Property' if x in property_crimes else 'Other')

# Group by community area and crime type
crime_by_community = df.groupby(['Community Area', 'Crime Type']).size().unstack()

# Plot for violent and property crimes in community areas
plt.figure(figsize=(10, 6))
crime_by_community.plot(kind='bar', stacked=True)
plt.title('Violent vs Property Crimes by Community Area')
plt.xlabel('Community Area')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Type')
plt.show()
```

#  Crime Type Analysis

```python
# Most Common Types of Crime by Primary Type and Description
# Count the number of crimes by 'Primary Type'
crime_by_type = df['Primary Type'].value_counts().head(10)

# Plot the top 10 most common crimes
plt.figure(figsize=(10, 6))
sns.barplot(x=crime_by_type.index, y=crime_by_type.values, palette='magma')
plt.title('Top 10 Most Common Types of Crime in Chicago')
plt.xlabel('Crime Type')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of crimes by 'Description' (secondary level)
crime_by_description = df['Description'].value_counts().head(10)

# Plot the top 10 most common crime descriptions
plt.figure(figsize=(10, 6))
sns.barplot(x=crime_by_description.index, y=crime_by_description.values, palette='coolwarm')
plt.title('Top 10 Most Common Crime Descriptions in Chicago')
plt.xlabel('Crime Description')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Comparing Categories of Crime (Violent, Property, Drug-Related) and Arrest Rates
# Define categories of crimes
violent_crimes = ['ASSAULT', 'BATTERY', 'HOMICIDE', 'ROBBERY']
property_crimes = ['BURGLARY', 'THEFT', 'MOTOR VEHICLE THEFT']
drug_related = ['NARCOTICS', 'DRUG EQUIPMENT VIOLATION']

# Create a new column categorizing crimes
def categorize_crime(primary_type):
    if primary_type in violent_crimes:
        return 'Violent'
    elif primary_type in property_crimes:
        return 'Property'
    elif primary_type in drug_related:
        return 'Drug-Related'
    else:
        return 'Other'

df['Crime Category'] = df['Primary Type'].apply(categorize_crime)

# Count the number of crimes in each category
crime_category_count = df['Crime Category'].value_counts()

# Plot the frequency of each category
plt.figure(figsize=(10, 6))
sns.barplot(x=crime_category_count.index, y=crime_category_count.values, palette='Set2')
plt.title('Crime Categories: Frequency in Chicago')
plt.xlabel('Crime Category')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# Arrest rates by crime category
arrest_rates_by_category = df.groupby('Crime Category')['Arrest'].mean()

# Plot the arrest rates by crime category
plt.figure(figsize=(10, 6))
sns.barplot(x=arrest_rates_by_category.index, y=arrest_rates_by_category.values, palette='dark')
plt.title('Arrest Rates by Crime Category')
plt.xlabel('Crime Category')
plt.ylabel('Arrest Rate')
plt.show()
```

```python
# Illinois Uniform Crime Reporting (IUCR) Codes with the Highest Number of Incidents
# Count the number of crimes by IUCR code
crime_by_iucr = df['IUCR'].value_counts().head(10)

# Plot the top 10 IUCR codes with the highest number of incidents
plt.figure(figsize=(10, 6))
sns.barplot(x=crime_by_iucr.index, y=crime_by_iucr.values, palette='Spectral')
plt.title('Top 10 IUCR Codes by Number of Incidents')
plt.xlabel('IUCR Code')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

# Arrest and Outcome

```python
# Percentage of Reported Incidents Resulting in Arrests
# Calculate overall arrest rate
arrest_rate = df['Arrest'].mean() * 100

print(f"Overall arrest rate: {arrest_rate:.2f}%")

# Plot the arrest rate
plt.figure(figsize=(6, 6))
plt.pie([arrest_rate, 100 - arrest_rate], labels=['Arrested', 'Not Arrested'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90)
plt.title('Percentage of Incidents Resulting in Arrests')
plt.show()
```

```python
# Arrest Likelihood for Different Types of Crimes
# Group by crime type and calculate arrest rates
arrest_rates_by_crime_type = df.groupby('Primary Type')['Arrest'].mean().sort_values(ascending=False).head(10)

# Plot the top 10 crimes with highest arrest rates
plt.figure(figsize=(10, 6))
sns.barplot(x=arrest_rates_by_crime_type.index, y=arrest_rates_by_crime_type.values, palette='Reds')
plt.title('Top 10 Crime Types with Highest Arrest Rates')
plt.xlabel('Crime Type')
plt.ylabel('Arrest Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
#Arrest Rates by Police District or Community Area
# Calculate arrest rates by district
arrest_rates_by_district = df.groupby('District')['Arrest'].mean().sort_values(ascending=False).head(10)

# Plot the top 10 police districts with the highest arrest rates
plt.figure(figsize=(10, 6))
sns.barplot(x=arrest_rates_by_district.index, y=arrest_rates_by_district.values, palette='Blues')
plt.title('Top 10 Police Districts with Highest Arrest Rates')
plt.xlabel('District')
plt.ylabel('Arrest Rate')
plt.show()
```

```python
# Calculate arrest rates by community area
arrest_rates_by_community = df.groupby('Community Area')['Arrest'].mean().sort_values(ascending=False).head(10)

# Plot the top 10 community areas with the highest arrest rates
plt.figure(figsize=(10, 6))
sns.barplot(x=arrest_rates_by_community.index, y=arrest_rates_by_community.values, palette='Purples')
plt.title('Top 10 Community Areas with Highest Arrest Rates')
plt.xlabel('Community Area')
plt.ylabel('Arrest Rate')
plt.show()
```

# Location Description Analysis

```python
# Most Common Locations for Crimes
# Count the number of crimes by location description
location_counts = df['Location Description'].value_counts().head(10)

# Plot the top 10 locations where crimes occur
plt.figure(figsize=(10, 6))
sns.barplot(x=location_counts.index, y=location_counts.values, palette='rocket')
plt.title('Top 10 Most Common Crime Locations in Chicago')
plt.xlabel('Location Description')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Specific Locations Prone to Particular Types of Crime
# Pivot table showing counts of crime by location and primary type
crime_by_location_type = df.pivot_table(index='Location Description', columns='Primary Type', aggfunc='size', fill_value=0)

# Filter for top 10 most common locations
top_locations = location_counts.index
crime_by_location_type_filtered = crime_by_location_type.loc[top_locations].T

# Plot a heatmap showing the types of crimes in top 10 locations
plt.figure(figsize=(12, 8))
sns.heatmap(crime_by_location_type_filtered, cmap='YlGnBu', annot=False)
plt.title('Heatmap of Crime Types in Common Locations')
plt.xlabel('Location')
plt.ylabel('Crime Type')
plt.show()
```

# Impact of External Factors

```python
# List of public holidays (just an example, adjust for specific holidays)
public_holidays = ['2020-01-01', '2020-12-25', '2020-11-26', '2020-07-04', '2020-05-25']

# Convert the Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create a new column indicating whether the crime occurred on a holiday
df['Holiday'] = df['Date'].dt.strftime('%Y-%m-%d').isin(public_holidays)

# Calculate crime rates on holidays and non-holidays
holiday_crime_rate = df.groupby('Holiday')['ID'].count()

# Plot the crime rates on holidays vs. non-holidays
plt.figure(figsize=(6, 6))
sns.barplot(x=holiday_crime_rate.index, y=holiday_crime_rate.values, palette='viridis')
plt.title('Crime Rates on Holidays vs Non-Holidays')
plt.xlabel('Holiday')
plt.ylabel('Number of Crimes')
plt.xticks([0, 1], ['Non-Holiday', 'Holiday'])
plt.show()
```

```python
# Crime Rates During COVID-19 Lockdown
 # Define the COVID-19 lockdown period (example: March 2020 - May 2020)
lockdown_start = '2020-03-20'
lockdown_end = '2020-05-31'

# Create a new column to identify lockdown period
df['Lockdown'] = df['Date'].between(lockdown_start, lockdown_end)

# Calculate crime rates during lockdown vs non-lockdown periods
lockdown_crime_rate = df.groupby('Lockdown')['ID'].count()

# Plot crime rates during lockdown vs non-lockdown periods
plt.figure(figsize=(6, 6))
sns.barplot(x=lockdown_crime_rate.index, y=lockdown_crime_rate.values, palette='coolwarm')
plt.title('Crime Rates During COVID-19 Lockdown vs Non-Lockdown')
plt.xlabel('Lockdown Period')
plt.ylabel('Number of Crimes')
plt.xticks([0, 1], ['Non-Lockdown', 'Lockdown'])
plt.show()
```

```python
# Analyzing Crime Rates by Hour of the Day
# Extract the hour of the day from the 'Date' column
df['Hour'] = df['Date'].dt.hour

# Group by hour and count the number of crimes
crime_by_hour = df.groupby('Hour')['ID'].count()

# Plot the number of crimes by hour of the day
plt.figure(figsize=(10, 6))
sns.lineplot(x=crime_by_hour.index, y=crime_by_hour.values, marker='o', color='orange')
plt.title('Crime Rates by Hour of the Day in Chicago')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Crimes')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

```python
# Analyzing Crime Rates by Day of the Week
# Extract the day of the week from the 'Date' column
df['Day of Week'] = df['Date'].dt.day_name()

# Group by day of the week and count the number of crimes
crime_by_day_of_week = df.groupby('Day of Week')['ID'].count()

# Re-order the days of the week for better visualization
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
crime_by_day_of_week = crime_by_day_of_week.reindex(days_order)

# Plot the number of crimes by day of the week
plt.figure(figsize=(10, 6))
sns.barplot(x=crime_by_day_of_week.index, y=crime_by_day_of_week.values, palette='muted')
plt.title('Crime Rates by Day of the Week in Chicago')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

# Spatial-Temporal Analysis


```python
# Extract hour and day of the week from the 'Date' column
df['Hour'] = df['Date'].dt.hour
df['Day of Week'] = df['Date'].dt.day_name()

# Create a pivot table with days of the week as rows and hours as columns
crime_by_day_hour = df.pivot_table(index='Day of Week', columns='Hour', aggfunc='size', fill_value=0)

# Re-order the days of the week for better visualization
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
crime_by_day_hour = crime_by_day_hour.reindex(days_order)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(crime_by_day_hour, cmap='coolwarm', annot=False, cbar=True)
plt.title('Heatmap of Crime Incidents by Day of the Week and Hour in Chicago')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()
```

```python
# Crime Patterns Between Night vs. Day
# Define day and night based on hours
df['Time of Day'] = df['Hour'].apply(lambda x: 'Day' if 6 <= x < 18 else 'Night')

# Group by time of day and count the number of crimes
crime_by_time_of_day = df.groupby('Time of Day')['ID'].count()

# Plot the crime patterns by time of day
plt.figure(figsize=(6, 6))
sns.barplot(x=crime_by_time_of_day.index, y=crime_by_time_of_day.values, palette='Set2')
plt.title('Crime Patterns: Day vs Night in Chicago')
plt.xlabel('Time of Day')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# Correlations Between Crime Types and Specific Geographic Areas
# Create a pivot table with districts as rows and primary types of crime as columns
crime_by_district_type = df.pivot_table(index='District', columns='Primary Type', aggfunc='size', fill_value=0)

# Plot a heatmap of crime types by district
plt.figure(figsize=(14, 8))
sns.heatmap(crime_by_district_type, cmap='YlGnBu', annot=False, cbar=True)
plt.title('Heatmap of Crime Types by Police District in Chicago')
plt.xlabel('Crime Type')
plt.ylabel('Police District')
plt.show()
```
