# Which cereals have the highest and lowest ratings?

```python
# the highest rating
highest_rated = df[['name', 'rating']].sort_values(by='rating', ascending=False).head(5)
print("Highest rated cereals:\n", highest_rated)

# the lowest rating
lowest_rated = df[['name', 'rating']].sort_values(by='rating', ascending=True).head(5)
print("Lowest rated cereals:\n", lowest_rated)
```

# What are the common nutritional patterns in highly rated cereals?

```python
threshold = df['rating'].quantile(0.75)
high_rated_cereals = df[df['rating'] > threshold]
high_rated_nutritional = high_rated_cereals[['calories', 'protein', 'fat', 'fiber', 'sugars']].mean()
print("Average nutritional values of highly rated cereals:\n", high_rated_nutritional)
```

# Which manufacturer produces the healthiest cereals based on nutrients?

```python
manufacturer_health = df.groupby('mfr')[['calories', 'fat', 'sugars', 'protein', 'fiber']].mean()
healthiest_manufacturers = manufacturer_health.sort_values(by=['sugars', 'fiber'], ascending=[True, False])
print("Healthiest manufacturers based on nutrients:\n", healthiest_manufacturers)
```

# What is the relationship between sugar content and cereal ratings?

```python
sns.scatterplot(x='sugars', y='rating', data=df)
plt.title("Relationship between Sugar Content and Ratings")
plt.show()

sugar_rating_corr = df['sugars'].corr(df['rating'])
print("Correlation between sugars and rating:", sugar_rating_corr)
```

# Which shelf placement tends to have higher-rated cereals?

```python
shelf_avg_rating = df.groupby('shelf')['rating'].mean().reset_index()

sns.barplot(x='shelf', y='rating', data=shelf_avg_rating)
plt.title("Average Rating by Shelf Placement")
plt.show()
```

# Are there significant differences between cold and hot cereals in terms of healthiness?

```python
type_nutritional = df.groupby('type')[['calories', 'protein', 'fat', 'fiber', 'sugars']].mean()
print("Nutritional comparison between cold and hot cereals:\n", type_nutritional)
type_nutritional.plot(kind='bar', figsize=(6,4))
plt.title("Nutritional Comparison between Cold and Hot Cereals")
plt.show()
```

# Do cereals with more vitamins tend to have higher ratings?

```python
vitamins_avg_rating = df.groupby('vitamins')['rating'].mean().reset_index()
sns.barplot(x='vitamins', y='rating', data=vitamins_avg_rating)
plt.title("Average Rating Based on Vitamin Content")
plt.show()
```

# What are the top cereals with the lowest calorie count per serving?

```python
low_calorie_cereals = df[['name', 'calories']].sort_values(by='calories').head(5)
print("Top 5 cereals with the lowest calorie count:\n", low_calorie_cereals)
```

 # Is there a correlation between sodium content and rating?

```python
sns.scatterplot(x='sodium', y='rating', data=df)
plt.title("Correlation between Sodium Content and Ratings")
plt.show()
sodium_rating_corr = df['sodium'].corr(df['rating'])
print("Correlation between sodium and rating:", sodium_rating_corr)
```

# Which cereals have the highest fiber content and what is their rating?

```python
high_fiber_cereals = df[['name', 'fiber', 'rating']].sort_values(by='fiber', ascending=False).head(5)
print("Top 5 high-fiber cereals and their ratings:\n", high_fiber_cereals)
```

# How does the average cereal rating vary across manufacturers?

```python
mfr_avg_rating = df.groupby('mfr')['rating'].mean().reset_index()
sns.barplot(x='mfr', y='rating', data=mfr_avg_rating, palette='coolwarm')
plt.title('Average Cereal Rating by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Rating')
plt.show()
```

# Which cereal types (cold vs. hot) are more popular in terms of ratings?

```python
type_avg_rating = df.groupby('type')['rating'].mean().reset_index()
sns.barplot(x='type', y='rating', data=type_avg_rating, palette='Set2')
plt.title('Average Rating for Cold vs. Hot Cereals')
plt.xlabel('Cereal Type')
plt.ylabel('Average Rating')
plt.show()
```

# What nutritional factors contribute most to high consumer ratings?

```python
corr_matrix = df[['rating', 'calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation between Rating and Nutritional Factors')
plt.show()
```

# What is the distribution of shelf placement, and how does it affect ratings?

```python
sns.countplot(x='shelf', data=df, palette='husl')
plt.title('Distribution of Cereals by Shelf Placement')
plt.xlabel('Shelf Level')
plt.ylabel('Count')
plt.show()
shelf_avg_rating = df.groupby('shelf')['rating'].mean().reset_index()
sns.barplot(x='shelf', y='rating', data=shelf_avg_rating, palette='viridis')
plt.title('Average Rating by Shelf Placement')
plt.xlabel('Shelf')
plt.ylabel('Average Rating')
plt.show()
```

# How do cereal ratings correlate with calories and sugar content?

```python
sns.scatterplot(x='calories', y='rating', data=df, color='red', label='Calories')
plt.title('Calories vs. Rating')
plt.xlabel('Calories per Serving')
plt.ylabel('Rating')
plt.show()

sns.scatterplot(x='sugars', y='rating', data=df, color='blue', label='Sugars')
plt.title('Sugars vs. Rating')
plt.xlabel('Sugars per Serving')
plt.ylabel('Rating')
plt.show()
```

 # Which cereals offer the best balance of nutrition and high ratings?

```python
balanced_cereals = df[(df['rating'] > 75) & (df['sugars'] < 10) & (df['fiber'] > 5)]
sns.barplot(x='name', y='rating', data=balanced_cereals, palette='Set1')
plt.title('Cereals with High Ratings and Balanced Nutrition')
plt.xlabel('Cereal name')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

 # What is the typical profile of a cereal with a rating above 75?

```python
high_rating_cereals = df[df['rating'] > 75]
nutritional_factors = ['calories', 'protein', 'fat', 'fiber', 'sugars']
high_rating_cereals[nutritional_factors].boxplot(figsize=(6,4))
plt.title('Nutritional Profile of Cereals with Ratings Above 75')
plt.show()
```

# What are the most common serving sizes for cereals, and how do they impact ratings?

```python
sns.histplot(df['weight'], kde=True, color='purple')
plt.title('Distribution of Serving Sizes (Weight)')
plt.xlabel('Weight (ounces)')
plt.show()
sns.scatterplot(x='weight', y='rating', data=df, color='orange')
plt.title('Serving Size vs. Rating')
plt.xlabel('Weight (ounces)')
plt.ylabel('Rating')
plt.show()

```

# Which cereals should be targeted for premium pricing based on their high ratings and nutritional value?

```python
premium_cereals = df[(df['rating'] > 80) & (df['fiber'] > 4) & (df['sugars'] < 8)]
sns.barplot(x='name', y='rating', data=premium_cereals, palette='magma')
plt.title('Cereals for Potential Premium Pricing')
plt.xlabel('Cereal Name')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

# How does shelf placement correlate with nutritional factors like sugar and fiber content?

```python
shelf_nutrition = df.groupby('shelf')[['sugars', 'fiber']].mean().reset_index()
shelf_nutrition.plot(kind='bar', x='shelf', figsize=(10,6))
plt.title('Average Sugar and Fiber Content by Shelf Placement')
plt.xlabel('Shelf')
plt.ylabel('Average Nutritional Content')
plt.show()
```

# Which manufacturers should focus on reformulating their cereals to improve ratings?

```python
# Group by manufacturer and calculate average ratings
mfr_avg_rating = df.groupby('mfr')['rating'].mean().reset_index()

# Filter manufacturers with below-average ratings
below_avg_mfr = mfr_avg_rating[mfr_avg_rating['rating'] < df['rating'].mean()]

# Bar plot for manufacturers with low ratings
sns.barplot(x='mfr', y='rating', data=below_avg_mfr, palette='Reds')
plt.title('Manufacturers with Low Average Ratings')
plt.xlabel('Manufacturer')
plt.ylabel('Average Rating')
plt.show()
```

#   What are the most underrated cereals based on nutritional content but lower ratings?

```python
# Filter cereals with good nutrition but higher sugar threshold and lower ratings
underrated_cereals = df[(df['fiber'] > 5) & (df['sugars'] < 10) & (df['rating'] < 60)]

# Check if there are any cereals that meet the conditions
if underrated_cereals.empty:
    print("No cereals found with the given criteria.")
else:
    # Bar plot to visualize underrated cereals
    plt.figure(figsize=(10, 6))  # Set the figure size
    sns.barplot(x='name', y='rating', data=underrated_cereals, palette='Blues_d')
    plt.title('Underrated Cereals with Good Nutrition')
    plt.xlabel('Cereal Name')
    plt.ylabel('Rating')
    plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to prevent clipping of tick labels
    plt.show()
```
