# Salary Trends Analysis:

```python
# Group by job_title and calculate the average salary in USD
average_salary_by_job_title = df.groupby('job_title')['salary_in_usd'].mean().reset_index()

# Sort by salary to get a better view
average_salary_by_job_title_sorted = average_salary_by_job_title.sort_values(by='salary_in_usd', ascending=False)

# Display the result
print(average_salary_by_job_title_sorted)
```

```python
# Salary Variation Across Different Experience Levels
# Group by experience_level and calculate the average salary in USD
average_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=average_salary_by_experience, x='experience_level', y='salary_in_usd', palette='rocket')
plt.title('Average Salaries by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.show()
```

```python
 # Salary Differences Between Employment Types
# Group by employment_type and calculate the average salary in USD
average_salary_by_employment_type = df.groupby('employment_type')['salary_in_usd'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=average_salary_by_employment_type, x='employment_type', y='salary_in_usd', palette='mako')
plt.title('Average Salaries by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Average Salary (USD)')
plt.show()
```

 # Remote Work and Compensation

```python
# Impact of Remote Work Ratio on Employee Salaries  
# Create a scatter plot to show the relationship between remote_ratio and salary_in_usd
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='remote_ratio', y='salary_in_usd', alpha=0.7)
plt.title('Impact of Remote Work Ratio on Employee Salaries')
plt.xlabel('Remote Work Ratio (%)')
plt.ylabel('Salary (USD)')
plt.axhline(y=df['salary_in_usd'].mean(), color='r', linestyle='--', label='Average Salary')
plt.legend()
plt.show()
```

```python
# Salary Comparison of Fully Remote vs. Hybrid/On-Site Roles
# Create a new column for remote work category
df['remote_category'] = df['remote_ratio'].apply(lambda x: 'Fully Remote' if x == 100 else ('Hybrid' if x > 0 else 'On-Site'))

# Group by remote_category and calculate the average salary in USD
average_salary_by_remote_category = df.groupby('remote_category')['salary_in_usd'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=average_salary_by_remote_category, x='remote_category', y='salary_in_usd', palette='plasma')
plt.title('Average Salaries by Remote Work Category')
plt.xlabel('Remote Work Category')
plt.ylabel('Average Salary (USD)')
plt.show()
```

```python
 #  Remote Work Ratios by Job Title or Experience Level Remote Work Ratios by Job Title or Experience Level
# Group by job_title and calculate the average remote_ratio
average_remote_by_job_title = df.groupby('job_title')['remote_ratio'].mean().reset_index()

# Sort by remote_ratio
average_remote_by_job_title_sorted = average_remote_by_job_title.sort_values(by='remote_ratio', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=average_remote_by_job_title_sorted, x='remote_ratio', y='job_title', palette='crest')
plt.title('Average Remote Work Ratio by Job Title')
plt.xlabel('Average Remote Ratio (%)')
plt.ylabel('Job Title')
plt.show()
```

```python
# Group by experience_level and calculate the average remote_ratio
average_remote_by_experience = df.groupby('experience_level')['remote_ratio'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=average_remote_by_experience, x='experience_level', y='remote_ratio', palette='cividis')
plt.title('Average Remote Work Ratio by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Remote Ratio (%)')
plt.show()
```

# Location-Based Analysis

```python
 #  Salaries Differ Based on Employee Residence and Company Location
# Create a box plot to compare salaries by employee residence and company location
plt.figure(figsize=(20, 10))
sns.boxplot(data=df, x='employee_residence', y='salary_in_usd', hue='company_location', palette='Set2')
plt.title('Salaries by Employee Residence and Company Location')
plt.xlabel('Employee Residence')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=45)
plt.legend(title='Company Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 2. Regions or Countries with the Highest Average Salaries
# Group by company_location and calculate the average salary in USD
average_salary_by_location = df.groupby('company_location')['salary_in_usd'].mean().reset_index()

# Sort by salary to get the highest average salaries
average_salary_by_location_sorted = average_salary_by_location.sort_values(by='salary_in_usd', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=average_salary_by_location_sorted.head(10), x='salary_in_usd', y='company_location', palette='coolwarm')
plt.title('Top 10 Regions with the Highest Average Salaries')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Company Location')
plt.show()
```

```python
#  Discrepancies Between Salaries in Employee Residence and Company Location for Remote Workers
# Filter for remote workers
remote_workers = df[df['remote_ratio'] == 100]

# Create a new DataFrame for discrepancies
discrepancies = remote_workers.groupby(['employee_residence', 'company_location'])['salary_in_usd'].mean().reset_index()

# Create a bar plot for discrepancies
plt.figure(figsize=(14, 7))
sns.barplot(data=discrepancies, x='employee_residence', y='salary_in_usd', hue='company_location', palette='muted')
plt.title('Salary Discrepancies Between Employee Residence and Company Location for Remote Workers')
plt.xlabel('Employee Residence')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.legend(title='Company Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Company Size and Salaries

```python
# Salary Distribution by Company Size
# Create a box plot to show salary distribution by company size
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='company_size', y='salary_in_usd', palette='pastel')
plt.title('Salary Distribution by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
 # Comparison of Salaries in Larger Companies vs. Job Roles or Experience Levels
sns.set(style='whitegrid')

# Create a histogram of salaries by company size
plt.figure(figsize=(14, 8))
sns.histplot(data=df, x='salary_in_usd', hue='company_size', multiple='stack', bins=30, kde=True)
plt.title('Distribution of Salaries by Company Size')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.legend(title='Company Size')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Remote Work Options Prevalence in Larger Companies
# Group by company_size and calculate the average remote_ratio
average_remote_by_company_size = df.groupby('company_size')['remote_ratio'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=average_remote_by_company_size, x='company_size', y='remote_ratio', palette='viridis')
plt.title('Average Remote Work Ratio by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Remote Ratio (%)')
plt.show()
```

# Employment Type Comparison

```python
# Create a box plot to compare salaries by employment type
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='employment_type', y='salary_in_usd', palette='Set3')
plt.title('Salaries by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
print(df['employment_type'].unique())
```

```python
# Job Titles Offered on Part-Time or Contract Basis
# Filter the DataFrame for part-time and contract employees
part_time_contract_df = df[df['employment_type'].isin(['PT', 'CT'])]

# Count occurrences of job titles
job_title_counts = part_time_contract_df['job_title'].value_counts().reset_index()
job_title_counts.columns = ['job_title', 'count']

# Check if job_title_counts is not empty before plotting
if not job_title_counts.empty:
    # Create a bar plot for job titles
    plt.figure(figsize=(12, 6))
    sns.barplot(data=job_title_counts.head(10), x='count', y='job_title', palette='magma')
    plt.title('Top 10 Job Titles Offered on Part-Time or Contract Basis')
    plt.xlabel('Count')
    plt.ylabel('Job Title')
    plt.show()
else:
    print("No job titles found for part-time or contract employment.")
```

```python
# Employment Type Influence on Remote Work Availability
# Group by employment type and calculate the average remote ratio
remote_by_employment_type = df.groupby('employment_type')['remote_ratio'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=remote_by_employment_type, x='employment_type', y='remote_ratio', palette='coolwarm')
plt.title('Average Remote Work Ratio by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Average Remote Ratio (%)')
plt.show()
```

# Predictive Analysis


```python
# Predicting Salary Based on Experience Level, Job Title, and Remote Ratio
# Create scatter plots to visualize relationships before modeling
plt.figure(figsize=(14, 5))

# Scatter plot for Experience Level vs Salary
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='experience_level', y='salary_in_usd', palette='Set2')
plt.title('Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')

# Scatter plot for Remote Ratio vs Salary
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='remote_ratio', y='salary_in_usd', alpha=0.6)
plt.title('Salary vs Remote Ratio')
plt.xlabel('Remote Ratio (%)')
plt.ylabel('Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
# Factors Influencing Salary Prediction
import numpy as np
# Check data types
print("Data types in the DataFrame:")
print(df.dtypes)

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

# Create a correlation matrix
correlation_matrix = numeric_df.corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#  Key Predictors of High Salary Levels for Certain Job Titles
# Group by job_title and calculate the average salary
average_salary_by_job_title = df.groupby('job_title')['salary_in_usd'].mean().reset_index()

# Sort to find high salary job titles
high_salary_job_titles = average_salary_by_job_title.sort_values(by='salary_in_usd', ascending=False)

# Create a bar plot for high salary job titles
plt.figure(figsize=(12, 6))
sns.barplot(data=high_salary_job_titles.head(10), x='salary_in_usd', y='job_title', palette='viridis')
plt.title('Top 10 Job Titles with Highest Average Salaries')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')
plt.show()
```

# Time-Based Trends

```python
# Salary Trends Over the Years in Different Industries
# Create a mapping of job titles to industries
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='work_year', y='salary_in_usd', estimator='mean', palette='tab10')
plt.title('Salary Trends Over the Years')
plt.xlabel('Year of Employment')
plt.ylabel('Average Salary (USD)')
plt.show()

plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='work_year', y='salary_in_usd', hue='experience_level', estimator='mean', palette='tab10')
plt.title('Salary Trends Over the Years by Experience Level')
plt.xlabel('Year of Employment')
plt.ylabel('Average Salary (USD)')
plt.legend(title='Experience Level')
plt.show()

```
